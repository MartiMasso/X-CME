```python
import heliosat
import datetime as dt
import numpy as np
import csv
import matplotlib.pyplot as plt
from heliosat.util import dt_utc_from_str
```

```python
t_s = dt_utc_from_str("2020-1-1")
t_e = dt_utc_from_str("2020-1-2")
```

```python
t_sta, b_sta = heliosat.Wind().get([t_s, t_e], "mag", as_endpoints=True, frame="HEEQ", return_datetimes=True)

# STA, WIND, SOLO, PSP  Wind (1995-Now); Parker (2018-Now), ...) Sometimes there are data gaps (maybe 1%). Discover (2018-Now)
```

```python
plt.figure(figsize=(18, 6))
plt.plot(t_sta, b_sta[:, 0], color="r")
plt.plot(t_sta, b_sta[:, 1], color="g")
plt.plot(t_sta, b_sta[:, 2], color="b")
plt.show()
```
